```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import io as io_url
```

```python
frame = io_url.imread('http://16385.courses.cs.cmu.edu/fall2020/assets/images/data/scotty.jpg')
frame = np.float32(frame) / 255

plt.imshow(frame)
```

```python
gray= cv2.cvtColor((255*frame).astype(np.uint8),cv2.COLOR_BGR2GRAY)
 
sift = cv2.SIFT_create()
kp = sift.detect(gray,None)
 
img=cv2.drawKeypoints(gray,kp,frame,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(img,cmap='gray');
```

```python
p,des = sift.compute(gray,kp)

print(des.shape)
print(des[0,:])
```
